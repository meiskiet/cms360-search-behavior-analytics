```python
# CREATE / SEED dim_content_rules

CATALOG = "workspace"
SCHEMA = "customer360"
DB = f"{CATALOG}.{SCHEMA}"
TABLE_RULES = f"{DB}.dim_content_new_rules"
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS {TABLE_RULES} (
    alias STRING,
    canonical_content STRING,
    category STRING
)
USING DELTA
""")

seed_rules = [
    ("truc tiep bong da", "trực tiếp bóng đá", "Sport"),
    ("bong da", "bóng đá", "Sport"),
    ("u19 viet nam", "u 19 việt nam", "Sport"),
    ("thu thach than tuong", "thử thách thần tượng", "Reality show"),
    ("running man", "running man", "Reality show"),
    ("vtv", "vtv", "News"),
    ("viet nam", "Việt Nam", "Sport"),
    ("bolero", "bolero", "Music"),
    ("penthouse", "penthouse", "K-Drama"),
    ("bat ma", "bắt ma phá án", "K-Drama"),
    ("liên minh công lý", "liên minh công lý: Phiên bản của Zack Snyder", "Action"),
    ("demon slayer kimetsu no yaiba", "lưỡi gươm diệt quỷ", "Animation"),
    ("2 ngay 1 dem", "2 ngày 1 đêm", "Reality show"),
    ("con tim sat da", "Con tim sắt đá", "Thai-Drama"),
    ("taxi driver", "taxi driver", "Action"),
    ("fpt play", "fpt play", "Sport")
]
```

```python
new_rules_df = spark.createDataFrame(seed_rules, ["alias", "canonical_content", "category"])

new_rules_df.write.mode("overwrite").saveAsTable(TABLE_RULES)

print("Rules table ready:", TABLE_RULES)
spark.table(TABLE_RULES).show(truncate=False)
```
